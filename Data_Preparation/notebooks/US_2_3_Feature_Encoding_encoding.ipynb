```python
import pandas as pd

df = pd.read_csv("Data_Preparation/data/processed/Dataset_ATS_v2_cleaned.csv")
```

```python

print(df['Contract'].unique())
print(df['Churn'].unique())
```

```python
contract_map = {
    'Month-to-month': 0,
    'One year': 1,
    'Two year': 2
}

#Churn (Target Variable)
#Defines churned customers as the positive class (1):

churn_map = {
    'No': 0,
    'Yes': 1
}

df['Contract'] = df['Contract'].map(contract_map)
df['Churn'] = df['Churn'].map(churn_map)


#Manual Encoding (Critical Variables)
#Manual mappings are applied first to override default encoders, which may sort alphabetically and introduce incorrect ordering.
```

```python
#Validation

df[['Contract', 'Churn']].isna().sum()
```

```python
#Column Check

print(df.columns.tolist())

```

```python
#4.2 Label Encoding (Binary Predictors)
#Binary features are encoded using categorical codes, resulting in values {0,1}.

binary_cols = [
    'gender',
    'Dependents',
    'PhoneService',
    'MultipleLines',
    'InternetService'
]
```

```python
for col in binary_cols:
    df[col] = df[col].astype('category').cat.codes
```

**Post-Encoding Validation**

```python
#Null Value Check

df.isna().sum()

#Spot-Check Semantic Correctness

df[['Contract', 'Churn']].value_counts().head()
```
